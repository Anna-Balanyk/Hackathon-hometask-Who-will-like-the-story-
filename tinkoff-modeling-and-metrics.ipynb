```python
#filling categorical and one numerical columns with '-1', age - with median age
#dropping unused columns
columns_to_neg = ['marital_status_cd', 'children_cnt', 'job_position_cd', 'gender_cd', 
                  'months_after_first_session']
cols_drop = ['first_ses', 'first_session_dttm', 'event_dttm',
            'minute', 'is_weekend', 'weekday', 'month', 'positive', 'skip']
full_data[columns_to_neg] = full_data[columns_to_neg].fillna(-1)
full_data['age']=full_data['age'].fillna(full_data['age'].median())
full_data = full_data.drop(cols_drop, axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
#splitting dataset to train and validation datasets
X, y = full_data.drop(columns=['event']), full_data['event']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=.4, random_state=10)
y_train.value_counts()
```

```python
test_data = pd.read_csv('/kaggle/input/merged-tinkoff/test_data.csv')
cus_tr = train_data.customer_id.unique().tolist()
cus_tes=test_data.customer_id.unique().tolist()
common = [i for i in cus_tr if i in cus_tes]
print('For train and test sets:', len(common), len(cus_tr), len(cus_tes), len(common)/len(cus_tes),"%")


c_train = X_train.customer_id.unique().tolist()
c_val= X_val.customer_id.unique().tolist()
comm = [i for i in c_train if i in c_val]
print('For X_train and X_valid sets:', len(comm), len(c_train), len(c_val), len(comm)/len(c_val), '%')
```

```python
#adding target columns to both datasets because we need them to do some processing on the train data
#excluding validation data
X_train['target'] = y_train
X_val['target'] = y_val 
```

```python
#parsing the labels, getting how many dislikes, skips, views and likes are there per user
sparse = pd.pivot_table(X_train, values = ['story_id'],
                                     index=['customer_id'], aggfunc='count',
                     columns=['target'], fill_value = 0).add_suffix('_user_gave')
sparse.head(10)
```

```python
#getting how many dislikes, skips, views and likes each story got
sparse2 = pd.pivot_table(X_train, values = ['customer_id'],
                                     index=['story_id'], aggfunc='count',
                     columns=['target'], fill_value = 0).add_suffix('_story_got')
sparse2.head(10)
```

```python
#merging datasets and dropping the targets
X_train = X_train.merge(sparse, on=['customer_id'])
X_train = X_train.merge(sparse2, on=['story_id'])

X_val = X_val.merge(sparse, on=['customer_id'])
X_val = X_val.merge(sparse2, on=['story_id'])

y_train = X_train['target']
y_val = X_val['target']
X_train.drop(['target'], axis=1, inplace=True)
X_val.drop(['target'], axis=1, inplace=True)
```

```python
#refreshing of label values and weights function
def label_to_id():
    return {
        'dislike': 1,
        'skip': 2, 
        'view': 3,
        'like': 4,
    }

def id_to_w():
    return {
        1: -10,
        2: -0.1,
        3: 0.1,
        4: 0.5,
    }

#writing a function for evaluating the results. 
#probs stands for probablitiies output of a model
#we multiply outputs with real weights and then divide this score on real weights
def score(probs, y_true):
    answers = probs
    id2w = id_to_w()
    true_weights = np.array(list(map(id2w.get, y_true)))
    my_score = (true_weights * answers).sum()
    true_score = np.abs(true_weights).sum()
    
    return my_score / true_score, my_score, true_score
```

```python
#basic RFC model, tuned by hand, checked max_depth till 8, but dropping this parameter gave better results
#n_estimators- checked 200, 300, 500, 1000
model = RandomForestClassifier(n_estimators=500, random_state=42)
model.fit(X_train, y_train)

val_probs = model.predict_proba(X_val)
weights = np.array([[-10, -0.1, 0.1, 0.5]])
weighted_ans = np.sum(val_probs * weights, axis=1)

score(weighted_ans, y_val)
```

```python

def postprocess_probs(probs, thresh):
    weights = np.array([[-10, -0.1, 0.1, 0.5]])
    probs = np.sum(probs * weights, axis=1)
    binarized = probs > thresh
    binarized = binarized * 2 - 1
    return binarized

def score(probs, y_true, thresh):
    answers = postprocess_probs(probs, thresh)
    id2w = id_to_w()
    true_weights = np.array(list(map(id2w.get, y_true)))
    our_score = (true_weights * answers).sum()
    best_score = np.abs(true_weights).sum()
    
    return our_score / best_score, our_score, best_score

sc = score(val_probs, y_val, 0)
sc
```

```python
weights = np.array([[-10, -0.1, 0.1, 0.5]]) #веса, которые надо использовать в метрике, 
#я использовала и для вычисления вероятности. То есть, каждую из четырех вероятностей умножила
#на соответствующий вес - дислайк - на (-10), и т.д. Итоговая сумма в строке - вероятность, которая 
#поможет сформировать бинаризованный ответ (-1 для дислайка и скипа, 1 - для просмотра и лайка.

probs = np.sum(val_probs * weights, axis=1) #v_probs - это вероятности, полученные после XGBoost. 
                                          #val_probs - вероятности после запуска RFC/   
probs
```

По поводу весов: они изрядно улучшают качество конечного результата. Я пробовала менять вес для дислайка и скипа, сначала это были крупные комбинации - -100 и -1, потом путем проб и ошибок вернулась к -0,1 для скипа и -10 для дислайка. Если выбирать значения, близкие к -10, общий результат стабильно ниже на 100-200 от общей суммы.

```python
#Бинаризую ответ. Если вероятность выше нуля - будет "1", ниже нуля - "-1".
thresh = 0

binarized = probs > thresh
binarized = binarized * 2 - 1
binarized[:100]
```

```python
def id_to_w():
    return {
        1: -10,
        2: -0.1,
        3: 0.1,
        4: 0.5,
    }
id2w = id_to_w()


true_weights = np.array(list(map(id2w.get, y_val)))
true_weights[:50]
```

в функции выше назначается вес каждого ответа (dislike - (-10), skip - (-0,1), view - (0.1), like - (0,5))
Дальше переведу ответы в вектор, то есть, вместо 1, 2, 3, 4 - будут стоять веса. Они помогут 
найти сумму реального ответа с учетом веса, где дислайк будет давать 10 баллов, лайк - 0,5 балла, 
скип - 0,1 балла и вью - тоже 0,1 балла.
Не страшно, что некоторые значения со знаком "минус". Сумма будет идти по модулю. Потому что если реальный ответ есть "-1" и вес такого ответа "-10", то получится "10". 

Итого, лучший возможный результат - 50518,2.

```python
best_score = np.abs(true_weights).sum()
best_score
```

Теперь надо посчитать результат, который выдала модель с учетом весов. Для этого возьму готовый вектор с возможными ответами binarized, там все ответы -1 или 1. Если умножить ответ "-1" на вес "-10" в случае правильного дислайка, можно получить +10, что хорошо отразится на общей сумме, но в случае неверного предсказания, результат будет -10. Пан или пропал.

Есть два вектора - один binarized, второй - true_weights. В нём уже собраны веса правильных ответов для валидационного датасета. Умножу их. Получу:
